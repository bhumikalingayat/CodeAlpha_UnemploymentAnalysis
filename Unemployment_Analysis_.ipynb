```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv("Unemployment in India.csv")

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df['date'] = pd.to_datetime(df['date'], dayfirst=True)
df = df.sort_values('date')
df['month'] = df['date'].dt.month_name()

plt.figure(figsize=(14,6))
sns.lineplot(x='date', y='estimated_unemployment_rate_(%)', data=df)
plt.title("ðŸ“ˆ India's Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()

covid_df = df[df['date'] >= '2020-03-01']
plt.figure(figsize=(14,6))
sns.lineplot(x='date', y='estimated_unemployment_rate_(%)', data=covid_df, color='red')
plt.title("ðŸ¦  COVID-19 Impact on Unemployment Rate")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

state_avg = df.groupby('region')['estimated_unemployment_rate_(%)'].mean().sort_values()
plt.figure(figsize=(12,10))
state_avg.plot(kind='barh', color='skyblue')
plt.title("ðŸ“Œ Average Unemployment Rate by State")
plt.xlabel("Unemployment Rate (%)")
plt.grid(axis='x')
plt.show()

monthly_avg = df.groupby('month')['estimated_unemployment_rate_(%)'].mean().sort_values()
plt.figure(figsize=(12,6))
monthly_avg.plot(kind='bar', color='orange')
plt.title("ðŸ“… Average Unemployment Rate by Month")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

print("âœ… Key Insights:")
print("- Unemployment rose sharply during COVID-19 lockdown (March 2020 onward).")
print("- States like Haryana, Rajasthan, and Delhi had higher average unemployment.")
print("- Seasonal trends show mid-year (Mayâ€“July) higher rates.")
print("- Useful for planning economic or social policies.")
```
